# 27 - Funciones: retorno de una lista

Hemos avanzado y visto que una función puede recibir como parámetros tipos de datos simples como enteros, flotantes etc. y estructuras de datos tipo lista.

También hemos visto que la función mediante la palabra clave return puede retornar un tipo de dato simple desde donde se la invocó.

Lo nuevo en este concepto es que una función también puede retornar una estructura de datos tipo lista. Con esto estamos logrando que una función retorne varios datos ya que una lista es una colección de datos.

## Ejemplos
### Ejemplo 1
Confeccionar una función que cargue por teclado una lista de 5 enteros y la retorne. Una segunda función debe recibir una lista y mostrar todos los valores mayores a 10. Desde el bloque principal del programa llamar a ambas funciones.

```python
# fuction enter list
# return a intger list of length n
def enter_list_integer():
    # empty list
    lista=[]
    # length list
    n=int (input("Enter the length of list: "))
    for f in range(n):
        lista.append(int(input(f"the  number {f+1} of {n} : ")))
    return lista

# function print greater than number
# param (lista, number)
def print_greater_number(lista, number):
    # take this with greater number
    greater=lista[0]
    for f in range(1,(len(lista))):
        if(lista[f]>greater):
            greater=lista[f]
    return greater

# main
lista=enter_list_integer()
print("List: ",lista)
num=10
print(f"greater number than {num} :",print_greater_number(lista,num))
```

### Ejemplo 3
Desarrollar un programa que permita cargar 5 nombres de personas y sus edades respectivas. Luego de realizar la carga por teclado de todos los datos imprimir los nombres de las personas mayores de edad (mayores o iguales a 18 años)
Imprimir la edad promedio de las personas.

```python
# function enter people data
# the name and age
def enter_people_data():
    #empty lists
    listName=[]
    listAge=[]
    # number of people
    n=int (input("Enter of number of people: "))
    for f in range(n):
        v1=input(f"Enter the person name {f+1} of {n}: ")
        listName.append(v1)
        v2=int(input(f"Enter the person age {f+1} of {n}: "))
        listAge.append(v2)
    return [listName,listAge]
    
# print of age
def print_people_of_age(name,age):
    for f in range (len(name)):
        if(age[f]>=18):
            print(name[f])

# average age
def average_age(age):
    average=0
    for f in range(len(age)):
        average+=age[f]
    average//=len(age)
    return average

# main 
[name, age]= enter_people_data()

print("People of age: ")
print_people_of_age(name,age)
print("*****************")
print("average of people: ",average_age(age))
```

 ### Problema 1
 En una empresa se almacenaron los sueldos de 10 personas.
Desarrollar las siguientes funciones y llamarlas desde el bloque principal:

 1) Carga de los sueldos en una lista.
 
 2) Impresión de todos los sueldos.
 
 3) Cuántos tienen un sueldo superior a 4000.
 
 4) Retornar el promedio de los sueldos.
 
 5) Mostrar todos los sueldos que están por debajo del promedio.

```python
# function salary of employee
# list with employees salary
# retunr listSalary
def employee_salary():
    # lista employee salary
    listSalary=[]
    # number of employee    
    n=int(input ("enter the number of employees: "))
    for f in range(n):
        # input data
        v1=int (input(f"Enter employee salary {f+1} of {n}: $"))
        # add salary
        listSalary.append(v1)
    return listSalary

# function print list
def employee_print_list(lista):
    for f in range(len(lista)):
        print(lista[f]) 
        
# count greater than number
# print employe greater than salary
# param list salary and salary 
# return counter grater
def employee_count_greaterThan(lista,number):
    countGreater=0
    for f in range(len(lista)):
        if(lista[f]>number):
            countGreater+=1
    return countGreater

# average of salary
# param lista salary
# return average 
def employee_salary_average(lista):
    average=0
    for f in range(len(lista)):
        average+=lista[f]
    return average/len(lista)

# print less than number
def employe_print_less_than(lista,number):
    for f in range(len(lista)):
        if(lista[f]<number):
            print(lista[f])

# main
print("Enter employee list: ")
employeeSalary=employee_salary()
print("**********************")
print("Employee list: ")
employee_print_list(employeeSalary)
print("**********************")
print("The number of employe greater than $4000: ", employee_count_greaterThan(employeeSalary,4000))
print("**********************")
employeAverage=employee_salary_average(employeeSalary)
print("Employe avarage salary: $", employeAverage)
print("**********************")
print("Employee less than average: ")
employe_print_less_than(employeeSalary,employeAverage)
```

### Problema 3
Confeccionar un programa que permita:

1) Cargar una lista de 10 elementos enteros.

2) Generar dos listas a partir de la primera. En una guardar los valores positivos y en otra los negativos.

3) Imprimir las dos listas generadas.

```python
# function enter numbers
def numbers_enter():
    # the amount of numbers
    n=int(input("Enter the amount of numbers: "))
    #list numbers
    listNumbers=[]
    # enter numbers
    for f in range(n):
        # input number
        num=float(input(f"Enter a number {f+1} of {n}: "))
        # add list
        listNumbers.append(num)
    # return list
    return listNumbers

# generate list numbers positive and negative
def numbers_list_positive_negative(listNumber):
    # list
    listNumbPositive=[]
    listNumberNegative=[]
    # sort out numbers
    for number in listNumber:
        #positive
        if(number>=0):
            listNumbPositive.append(number)
        else:
            listNumberNegative.append(number)
    return listNumbPositive,listNumberNegative
# main
# enter number list
listNumb=numbers_enter()
[listPos, listNeg]=numbers_list_positive_negative(listNumb)
print("List numbers Positive:",listPos)
print("List numbers Negative:",listNeg)
```
