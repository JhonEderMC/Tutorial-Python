# 36 - Funciones: parámetros mutables e inmutables
En Python tenemos tipos de datos inmutables:

`
enteros
float
string
tuplas
`

Mutables:

`
listas
diccionarioso 
`

Esto tiene mucha importancia cuando enviamos a una función una variable mutable, veremos con un ejemplo como podemos pasar como parámetro una lista a una función y posteriormente cambiar su contenido y esto se vea reflejado en la variable que le enviamos al llamarla.

## Ejemplos
### Ejemplo 1
Confeccionar un programa que contenga las siguientes funciones:
1. Carga de una lista y retorno al bloque principal.
2. Fijar en cero todos los elementos de la lista que tengan un valor menor a 10.
3. Imprimir la lista

```python
# enter list
def number_create_list():
    # list of numbers
    numberList=[]
    stop=False
    # enter list
    while(stop!=True):
        number=int(input("Enter a number: "))
        # add to list
        numberList.append(number)
        cond=input("do you want enter another number?[y/n]: ")
        # stop
        if(cond.lower()=='n' or cond.lower=="no"):
            stop=True
    return numberList

# function zero all numbers greater less than 10
def number_zero_less_than(numberList):    
    for f in range(len(numberList)):
        if (numberList[f]<10):
            numberList[f]=0

# function print list
def number_print(numberList):
    for number in numberList:
        print(number,end=" ")

#main
number=number_create_list()
print("*"*35)
number_zero_less_than(number)
number_print(number)
        
```

La primer función permite la carga de una lista de enteros hasta que el operador no desee cargar más valores:

`

Lo nuevo aparece en la función number_zero_less_than que recibe como parámetro una lista llamada "numberList" y dentro de la función modificamos los elementos de la lista, estos cambios luego se ven reflejados en la variable definida en el bloque principal de nuestro programa:

`

Si ejecutamos este programa e ingresamos algunos elementos de la lista con valores inferiores a 10 veremos luego que la variable global "number" es modificada

**Nota:** en conclusión, hay que tener cuidado cuando le pasamos una variable modificable a una funcion, porque si esta es modificada la variable original también.

### Ejemplo 2
Confeccionar un programa que contenga las siguientes funciones:
1. Carga de una lista de 5 nombres.
2. Ordenar alfabéticamente la lista.
3. Imprimir la lista de nombres            

```python
# function enter list names
# return list numbers
def name_enter_list():
    #empty list
    listName=[]
    i=0
    while i<5:
        name=input(f"Enter person's name{i+1} of 5: ")
        listName.append(name)
        i+=1
    return listName

# function alphabetical ordered
def name_ordered(nm):
    # n-1 iterations
    for k in range(4):
        #optimization, each step of the external for is 
        #equivalent to a word organized alphabetically
        for f in range(len(nm)-1-k):
            # if it is greater
            if (nm[f]>nm[f+1]): 
                auxNm=nm[f]
                nm[f]=nm[f+1]
                nm[f+1]=auxNm

 # print function
def name_print(nm):
    for n in nm:
        print(n,"",sep=" ",end="")
    print()
    
        

# main
nam=name_enter_list()
print("*"*25)
print("Name  before: ")
name_print(nam)
print("*"*25)
print("Name  after: ")
name_ordered(nam)
name_print(nam)
        
```

### Ejemplo 3
Confeccionar un programa que almacene en un diccionario como clave el nombre de un contacto y como valor su número telefónico:
1. Carga de contactos y su número telefónico.
2. Pemitir modificar el número telefónico. Se ingresa el nombre del contacto para su búsqueda.
3. Imprimir la lista completa de contactos con sus números telefónicos.

```python
# function create dictionary phone number
# format: key-> name value-> phone number
def phone_create_data():
    # dictionary phone numbers empty
    dictPhone={}
    stop=False
    while(stop!=True):
        # enter data
        name=input("Enter person name: ")
        number=int(input(f"Enter {name} phone number: "))
        # add to dictionary
        dictPhone[name]=number
        # do you want continue
        cond=input("Do you want enter another number?[y/n]: ")
        # condition
        if(cond.lower()=='n' or cond.lower()=="no"):
            # exit while
            stop=True 
    return dictPhone

# function modify a phone number
# function modify the dictionary
def phone_change_phone(dictPh):
    stop=False
    while(stop!=True):
        # name to search
        name=input("Enter a person's name to seach: ")
        # it if it exists in the dictionary
        if name in dictPh:
            # new phone number
            phone=int(input("Enter new phone number: "))
            # change number in dictionary
            dictPh[name]=phone
        else:
            print("The number was not found")
        cond=input("Do you want search another name?[y/n]: ")
        # condition
        if(cond.lower()=='n' or cond.lower()=="no"):
            # exit while
            stop=True 

# function print to list (name phone)
def phone_print_numbers(dictPh):
    for name in dictPh:
        print("Name",name,"Phone",dictPh[name],sep=": ")

        
#main
phone=phone_create_data()
print("*"*35)
print("Print directory: ")
phone_print_numbers(phone)
print("*"*35)
phone_change_phone(phone)
print("Print directory: ")
phone_print_numbers(phone)
```

La primer función nos permite cargar los nombres de contactos con sus respectivos teléfonos en un diccionario llamado "contactos", la función retorna este diccionario y se almacena posteriormente en una variable global

La función "modificar_telefono" recibe como parámetro el diccionario "contactos", solicita la carga de un contacto, en el caso que exita se procede a solicitar el nuevo teléfono y modificar el diccionario

*Con la modificación del parámetro en la función "modificar_telefono" se está modificando la variable global que le pasamos desde el bloque principal* 

### Problema 1

Crear un diccionario en Python para almacenar los datos de empleados de una empresa. La clave será su número de legajo y en su valor almacenar una lista con el nombre, profesión y sueldo.
Desarrollar las siguientes funciones:
1. Carga de datos de empleados.
2. Permitir modificar el sueldo de un empleado. Ingresamos su número de legajo para buscarlo.
3. Mostrar todos los datos de empleados que tienen una profesión de "analista de sistemas"

```python
##### function dictionary employees
# key-> ID values-> name,profession,salary
# return dictionary
def employee_create_dictionary():
    # empty dictionary
    dictEmployee={}
    # number of employees
    n=int(input("Number of employees: "))
    # enter employee data
    for f in range(n):
        # list employee data
        listData=[]
        name=input(f"Enter employee name{f+1} of {n}: ")
        #add to list
        listData.append(name)
        ID=int(input(f"Enter {name} ID: "))
        profession=input(f"Enter {name} profession: ")
        listData.append(profession)
        salary=int(input(f"Enter {name} Salary: $"))
        listData.append(salary)
        # add to dictionary       
        dictEmployee[ID]=listData
    return dictEmployee

# function change salary
def employee_change_salary(dictEmp):
    stop=False
    while(stop!=True):
        # id to search
        ID=int(input("Enter employee ID to search: "))
        # if ID exist in the dictionary
        if ID in dictEmp:
            print("old salary: $",dictEmp[ID][2])
            newSalary=int(input("Enter the new salary: $"))
            dictEmp[ID][2]=newSalary
        else:
            print("ID was no found")
        # condition to stop
        cond=input("Do you want change another salary?[y/n]: ")
        if (cond.lower()=='n' or cond.lower()=="no"):
            stop=True
    
# print employee "analistas de sistemas"
def employee_print_analista_sistemas(dictEmp):
    print("Profesion analista de sistemas: ")
    for ID in dictEmp:
        if (dictEmp[ID][1].lower()=="analista de sistemas"):
            print("Name",dictEmp[ID][0],"Salary",dictEmp[ID][2],sep=": ")
            
            

# main
employee=employee_create_dictionary()
print("*"*35)
employee_change_salary(employee)
print("*"*35)
employee_print_analista_sistemas(employee)         
```
