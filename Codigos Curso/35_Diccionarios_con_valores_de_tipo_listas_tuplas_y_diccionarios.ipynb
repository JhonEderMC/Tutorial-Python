# 35 - Diccionarios: con valores de tipo listas, tuplas y diccionarios
**Lo más poderoso que podemos encontrar en las estructuras de datos en Python es que podemos definir elementos que sean también estructuras de datos.** En general se dice que podemos anidar una estructura de datos dentro de otra estructura de datos.

Ya vimos en conceptos anteriores que podemos definir elementos de una lista que sean también de tipo lista o de tipo tupla.

Hemos dicho que un diccionario consta de claves y valores para esas claves. Desarrollaremos problemas donde los valores para esas claves sean tuplas y o listas.

### Ejemplo 1
Confeccionar un programa que permita cargar un código de producto como clave en un diccionario. Guardar para dicha clave el nombre del producto, su precio y cantidad en stock.
Implementar las siguientes actividades:

2. Listado completo de productos.
3. Consulta de un producto por su clave, mostrar el nombre, precio y stock.

```python
""""
function creater dictionary products stock
return dictProducts
form: key->code, values-> name, price, code
"""
def product_create_dictProducts():
    #empy dictionary 
    dictProducts={}
    # number of products
    n=int(input("Enter the number of products: "))
    # enter products
    for f in range(n):
        # data
        name=input(f"Enter porduct's name {f+1} of {n}: ")
        code=int(input(f"Enter {name}'s code: "))
        price=int(input(f"Enter {name}'s price: $")) 
        stock=int(input(f"Enter {name}'s stock: "))
        # add dictionary
        dictProducts[code]=[name,price,stock]
    return dictProducts

"""
function print dictionary of products
param dictionary of products
search is diferent to zero, search codi

"""
def product_print_dictionary(dictProducts, search=0):
    if (search==0):
        for code in dictProducts:
            print("Code",code," Name",dictProducts[code][0]," Price",dictProducts[code][1]," Stock",dictProducts[code][2],sep=": ")
    else:
        stop=False
        while(stop!=True):
            code=int(input ("Enter to code to search: "))
            # find
            if code in dictProducts:                
                # it was found
                print("Name",dictProducts[code][0], "Price",dictProducts[code][1]," Stock",dictProducts[code][2],sep=": ")
            # it was not found
            else:
                print("the code was not found")
            s=input("Do you want searh another product? [y/n]: ")
            if(s.lower()=="n" or s.lower()=="no"):
                stop=True

"""
function print products with stock cero
"""
def product_print_stock_zero(dictProducts):                          
    for code in dictProducts:
        #check stock
        if(dictProducts[code][2]==0):
            print("Code",code,"Name",dictProducts[code][0])

                      
# main 
#create dictionary
dictProducts=product_create_dictProducts()
print("*"*35)
# print data
print("Stock data: ")
product_print_dictionary(dictProducts)
print("*"*35)
# search
product_print_dictionary(dictProducts,2) 
print("*"*35)
# stock zero
product_print_stock_zero(dictProducts)                  
                      
                      
```

### Ejemplo 2
Confeccionar una agenda. Utilizar un diccionario cuya clave sea la fecha. Permitir almacenar distintas actividades para la misma fecha (se ingresa la hora y la actividad)
Implementar las siguientes funciones:
1. Carga de datos en la agenda.
2. Listado completo de la agenda.
3. Consulta de una fecha.

```python
# function create agenda
# kay date
def agenda_create():
    #dictionary
    dictAgenda={}
    stopAgenda=False
    # enter information
    while(stopAgenda!=True):
        # enter date
        date=input("Enter a date in format (dd/mm/yyy): ")
        # list of activities this date               
        listDate= []
        # this date
        stopDate=False
        while (stopDate!=True):
            # enter data
            hour=int(input("Enter hour: "))
            activity=input("Enter activity: ")
            listDate.append((hour,activity))
            # continue or no add other activity
            cont=input(f"Do you want enter another activity this day ({date})?[y/n]: ")
            if (cont.lower=='n' or cont.lower()=="no"):
                stopDate=True
        #add dictionary
        dictAgenda[date]=listDate
        #continue or no add other date (agend)
        cont=''
        cont=input(f"Do you want enter another date?[y/n]: ")
        if (cont.lower=='n' or cont.lower()=="no"):
            stopAgenda=True
    # return
    return dictAgenda        
                
 
# funtion print full agenda
def agenda_print (dictAgenda):
    for date in dictAgenda:
        print("Date:",date)
        # hour and activty
        for f in range(len(dictAgenda[date])):
            print("Hour",dictAgenda[date][f][0],"Activity",dictAgenda[date][f][1],sep=": ")

# function search date
def agenda_search_date(dictAgenda):
    stop=False
    while(stop!=True):
        search=input("Enter a date to search (dd/mm/yyy): ")
        if search in dictAgenda:
            for hour,activity in dictAgenda[search]:
                print("hour",hour,"activity",activity,sep=": ")
        else:
            print("no found")            
        cont=input(f"Do you want search another date?[y/n]: ")
        if (cont.lower=='n' or cont.lower()=="no"):
            stop=True
    

# main
# create agenda
dictAgenda=agenda_create()
print("*"*40)
# print agenda
agenda_print(dictAgenda)
print("*"*40)
# search date
agenda_search_date(dictAgenda)

```
